```python
"""
############################
Genrated from Chatgpt
############################
"""
import os
from PIL import Image
import numpy as np

folder = "datasets1/carplate1/train/images"
sizes = []

for img_file in os.listdir(folder):
    if img_file.endswith(('.jpg', '.png', '.jpeg')):
        img = Image.open(os.path.join(folder, img_file))
        sizes.append(img.size)  # (width, height)

widths, heights = zip(*sizes)
print("Total images:", len(sizes))
print("Min size:", (min(widths), min(heights)))
print("Max size:", (max(widths), max(heights)))
print("Average size:", (int(np.mean(widths)), int(np.mean(heights))))
```

Model1 evaluation (without data of car in the night)

```python
model = YOLO("runs/detect/train/weights/best.pt")

results = model.val(data="datasets1/carplate1/data.yaml", imgsz=640, batch=16)

print("Precision per class:", results.box.p)
print("Recall per class:", results.box.r)
print("mAP50:", results.box.map50)
print("mAP50-95:", results.box.map)
```

Model2 evaluation (with data of car in the night)

```python
model2 = YOLO("runs/detect/train2/weights/best.pt")

results = model2.val(data="datasets2/carplate2_added/data.yaml", imgsz=1280, batch=16)

print("Precision per class:", results.box.p)
print("Recall per class:", results.box.r)
print("mAP50:", results.box.map50)
print("mAP50-95:", results.box.map)
```
